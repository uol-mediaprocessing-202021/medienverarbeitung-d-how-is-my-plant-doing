# Bing image downloader
In this notebook the bing image downloader (https://github.com/gurugaurav/bing_image_downloader) is used for creating a data base for our training data. 

```python
from bing_image_downloader import downloader
```

```python
base_path_img = './dataset/'
```

```python
# list with search strings for bing images
list_search_strings = ['apple scab leaf', 'apple leaf', 'apple rust leaf', 
                       'bell pepper leaf spot', 'bell pepper leaf', 
                       'corn gray leaf spot', 'corn leaf blight', 'corn rust leaf',
                       'peach leaf', 'peach leaf curl', 'peach leaf bacterial spot', 
                       'potato leaf', 'potato leaf blight', 
                       'raspberry leaf', 'raspberry leaf spot', 
                       'soybean leaf', 'soybean leaf rust', 'soybean leaf blight'
                       'squash leaf', 'squash powdery mildew leaf', 
                       'strawberry leaf', 'strawberry rust leaf', 
                       'tomato leaf', 'tomato leaf blight', 'tomato leaf spot', 'tomato leaf mosaic virus', 
                       'tomato leaf yellow virus', 'tomato mold leaf', 
                       'grape leaf', 'grape leaf black rot']
```

```python
# limit: nunmber of max downloaded images per search string, timeout: max time per request
for search_string in list_search_strings:
    downloader.download(search_string, limit=1500,  output_dir=base_path_img, adult_filter_off=False, force_replace=False, timeout=2)
```
